```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('DataSets/usuarios.csv')
data.head(20)
```

```python
data.isnull().sum()
```

```python
_list = [x for x in data.columns if x != 'clase']
```

```python
x = np.array(data[_list])
y = np.array(data[['clase']])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(x_train, y_train.ravel())
```

```python
y_pred = log.predict(x_test)
```

```python
y_pred
```

```python
log.intercept_, log.coef_
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred)) # Mdelo malísimo por culpa de la falta de datos
```
