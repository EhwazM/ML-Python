```python
data3 = pd.read_csv('DataSets/titanic/titanic3.csv')
data3 = data3.fillna('Unkown')
```

```python
data3.iloc[1302:1309]
```

```python
data4 = pd.read_csv('DataSets/titanic/titanic3.csv')
data4['age'] = data4['age'].fillna(data4['age'].mean())
data4.iloc[1302:1309]
```

```python
data5 = pd.read_csv('DataSets/titanic/titanic3.csv')
data5['age'] = data5['age'].fillna(method = 'ffill')
data5.iloc[1302:1309]
```

```python
data6 = pd.read_csv('DataSets/titanic/titanic3.csv')
data6['age'] = data6['age'].fillna(method = 'bfill')
data6.iloc[1302:1309]
```
