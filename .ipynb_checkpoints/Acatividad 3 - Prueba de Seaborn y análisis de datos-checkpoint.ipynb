```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('https://datos.gov.co/resource/y628-5q9a.csv')
data.head(10)
```

```python
data.info()
```

```python
data['clase_accidente'].value_counts().plot(kind = 'pie', autopct = '%1.1f%%')
```

```python
cross_table = pd.crosstab(data['clase_accidente'], data['gravedad_accidente'])
cross_table
cross_table.plot(kind = 'bar')
plt.show()
```

```python
drivers = data[(data['condicion_victima'] == 'Conductor') | (data['condicion_victima'] == 'Motociclista')]
tabla_cruzada2 = pd.crosstab(drivers['edad_victima'], drivers['clase_accidente'])
tabla_cruzada2.plot(kind = 'bar')
plt.show()
```

### Hay una cantidad notable de j坦venes conductores presentes en alg炭n tipo de accidente, mayormente en choques. ###

```python
tabla_cruzada3 = pd.crosstab(drivers['sexo_victima'], drivers['condicion_victima'])
tabla_cruzada3.plot(kind = 'bar')
plt.show()
```
